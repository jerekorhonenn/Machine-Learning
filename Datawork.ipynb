```python
data2=data[["Onluokka", "Osallkm", "Valoisuus","Temperature", "Vakavuusko"]]
```

```python
data2
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(np.abs(data2.corr()), annot=True, cmap="viridis", fmt="0.2f");
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
```

```python
X, y= data2.values[:,:-1], data2.values[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, stratify=y)
```

```python
clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(8,2), random_state=1,max_iter=1000)

clf.fit(X_train, y_train)
```

```python
clfaccuracy = clf.score(X_train, y_train)
print("accuracy = ", clfaccuracy*100, "%")
```

```python
test_accident=[[3, 1, 3,6]]
#Now I tried to find an accident parameters an 3, 1, 3 and 6 it should cause 2 and it means injury but not deadly accident
# first 3 is ohitus accident, 1 is number of people, 3 is the light and 3 is dark without lights and 6 celsius is the temperature
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
accuracy = model.score(X_train, y_train)
print("accuracy = ", accuracy*100, "%")
```

```python
coeff=list(model.coef_[0])
labels = data2.columns[:-1]
features = pd.DataFrame()
features['Features'] = labels
features['importance'] = coeff
features.sort_values(by=['importance'], ascending=True, inplace=True)
features['positive'] = features['importance'] > 0
features.set_index('Features', inplace= True)
features.importance.plot(kind='barh', figsize =(11, 6), color= features.positive.map({True: 'blue', False: 'red'}))
plt.xlabel('importance')

```
